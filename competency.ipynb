```python
from pymongo import MongoClient
from pymongo import GEOSPHERE
import pandas as pd
import requests
import json
from dotenv import load_dotenv
import os
```

```python
client = MongoClient("localhost:27017")
db = client.get_database("ironhack")
companies = db.get_collection("companies")
offices = client.companies.offices
```

```python
res = list(offices.find({},{"offices":1}))
```

```python
query = {'$and': [{'funding_rounds.raised_amount' : {'$gt': 1000000}},
                  {"offices.country_code": 'ESP'}, {"offices.city": 'Madrid'}]}
```

```python
selected_offices = list(offices.find(query,{"_id":0,'offices': 1, "name": 1}))
```

```python
selected_offices
```

```python
df_competency = pd.DataFrame(selected_offices)
```

```python
df_competency
```

```python
latitude = []
for x in range(0, len(df_competency)):
       latitude.append(df_competency.offices[x]['latitude'])
```

```python
df_competency['latitude'] = pd.DataFrame(latitude)
```

```python
longitude = []
for x in range(0, len(df_competency)):
       longitude.append(df_competency.offices[x]['longitude'])
```

```python
df_competency['longitude'] = pd.DataFrame(longitude)
```

```python
df_competency
```

```python
df_competency = df_competency.drop(['offices'], axis=1)
```

```python
df_competency
```

```python
df_competency.to_csv('data/competency.csv', index=False)
```
